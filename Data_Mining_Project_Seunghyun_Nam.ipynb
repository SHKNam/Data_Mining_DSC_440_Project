**Logistic Regression**
- This model is ideal for binary classification problems. As 'SAP' is a binary outcome (presence or absence of a condition), logistic regression can model the probability of occurrence based on the input features.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import make_scorer, accuracy_score, f1_score
from sklearn.model_selection import StratifiedKFold, cross_validate
import numpy as np

# df = pd.read_csv('your_data.csv'); # Replace 'your_data.csv' with your actual data file

Feature_data = df[['AGE', 'EDUC', 'ETHNIC', 'RACE', 'GENDER', 'SPHSERVICE',
                    'CMPSERVICE', 'OPISERVICE', 'RTCSERVICE', 'IJSSERVICE', 'SUB', 'MARSTAT', 'SMISED',
                    'SAP', 'EMPLOY', 'DETNLF', 'VETERAN', 'LIVARAG', 'NUMMHS',
                    'STATEFIP', 'DIVISION', 'REGION']]

Target_data = df[['TRAUSTREFLG', 'ANXIETYFLG', 'ADHDFLG', 'CONDUCTFLG', 'DELIRDEMFLG',
                  'BIPOLARFLG', 'DEPRESSFLG', 'ODDFLG', 'PDDFLG', 'PERSONFLG', 'SCHIZOFLG',
                  'ALCSUBFLG', 'OTHERDISFLG']]

Feature_data = Feature_data.fillna(0)
Target_data = Target_data.fillna(0)

for column in Target_data:
    X = Feature_data
    y = Target_data[column]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)

    # Logistic Regression Model
    model = LogisticRegression()

    # Stratified K-Fold for cross-validation
    stratified_kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=20)

    # Scoring metrics
    scoring = {'accuracy': make_scorer(accuracy_score), 'f1': make_scorer(f1_score)}

    # Perform cross-validation
    cv_results = cross_validate(model, X, y, cv=stratified_kfold, scoring=scoring)

    mean_accuracy = np.mean(cv_results['test_accuracy'])
    mean_f1_score = np.mean(cv_results['test_f1'])

    print(f'Target: {column}')
    print(f'Mean Accuracy: {mean_accuracy:.4f}')
    print(f'Mean F1 Score: {mean_f1_score:.4f}\n')
```
